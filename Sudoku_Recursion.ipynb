##SAMPLE PROBLEM

image.png

```
GRID=np.array([[5, 0, 9,   0, 0, 0,  0, 0, 7],
               [0, 8, 0,   0, 1, 0,  5, 2, 0],
               [0, 0, 3,   0, 8, 4,  0, 0, 1],

               [0, 9, 0,   7, 0, 0,  0, 0, 2],
               [4, 0, 0,   0, 5, 0,  3, 9, 0],
               [8, 0, 2,   1, 0, 0,  0, 0, 4],

               [0, 0, 0,   3, 0, 2,  0, 0, 5],
               [0, 4, 0,   0, 0, 0,  7, 0, 0],
               [1, 0, 7,   0, 9, 0,  0, 8, 0]], dtype = int)
```

##HELPER FUNCTIONS

```
def get_box(x=0,y=0):
  """get 3X3 box of the cell"""
  x_start = 3*(x//3)
  y_start = 3*(y//3)
  return GRID[x_start:x_start+3,y_start:y_start+3]

def pretty_print():
  """function to print grid"""
  for i in range(9):    
    for j in range(9):
      print(GRID[i,j],end = '|' if j%3 == 2 else ' ')
    if i%3 == 2:
      print()
    print()

def check_presence(n,m):
  """check if value in matrix"""
  return n in m.flatten()

def is_possible(num,x,y):
    """check if move is possible"""
    #check horizontally
    if check_presence(num,GRID[x]):
      #print(1)
      return False

    #check vertically
    if check_presence(num,GRID[:,y]):
      #print(2)
      return False

    #check in 3X3 box
    if check_presence(num,get_box(x,y)):
      #print(3)
      return False

    return True
```

##SOLVE BY RECURSION

```
def solve():
  global GRID
  for x in range(9):
    for y in range(9):
      if GRID[x,y] == 0:
        for num in range(1,10):
          if is_possible(num,x,y):
            GRID[x,y] = num
            solve()
            GRID[x,y] = 0
        return
  print("Solved")
  pretty_print()
```

##SOLUTIONS

```
 solve()
```
